```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plotantena(punto: int, antena: int, con_elevacion: bool, intento: int, desde_segundo: int):
    data = pd.read_csv(f'./data_log/punto_{punto}_antena_{antena}_{"con" if con_elevacion else "sin"}_elevacion_intento_{intento}.csv')
    data_A = data[data['id'] == '6C1DEBAFB3B5']
    data_B = data[data['id'] == '6C1DEBAFB644']

    newer_A = data_A[data_A['tiempo'] > desde_segundo]

    plt.xlim(-45, 45)

    if con_elevacion:
        plt.ylim(-45, 45)
    else:
        plt.ylim(0, 1)

    plt.title(f'6C1DEBAFB3B5 punto {punto} antena {antena} {"con" if con_elevacion else "sin"} elevacion intento {intento} desde el segundo {desde_segundo}')
    plt.scatter(newer_A['azimuth'], newer_A['elevation'] if con_elevacion else np.random.rand(len(newer_A)), c=newer_A['tiempo'])
    # plot average
    averages = [np.average(newer_A['azimuth']), np.average(newer_A['elevation'] if con_elevacion else np.random.rand(len(newer_A)))]
    print("promedios: ", averages)
    plt.scatter(averages[0], averages[1], marker="*", c='red')
    plt.show()

    # print(np.std(newer_A['azimuth']), np.std(newer_A['elevation']))

    newer_B = data_B[data_B['tiempo'] > desde_segundo]

    plt.xlim(-45, 45)
    if con_elevacion:
        plt.ylim(-45, 45)
    else:
        plt.ylim(0, 1)

    plt.title(f'6C1DEBAFB644 punto {punto} antena {antena} {"con" if con_elevacion else "sin"} elevacion intento {intento} desde el segundo {desde_segundo}')
    plt.scatter(newer_B['azimuth'], newer_B['elevation'] if con_elevacion else np.random.rand(len(newer_B)), c=newer_B['tiempo'])

    averages = [np.average(newer_B['azimuth']), np.average(newer_B['elevation'] if con_elevacion else np.random.rand(len(newer_B)))]
    print("promedios: ", averages)
    plt.scatter(averages[0], averages[1], marker="*", c='red')
    plt.show()

    # print(np.std(newer_B['azimuth']), np.std(newer_B['elevation']))
```

```python
plotantena(punto=2, antena=2, con_elevacion=True, intento=2, desde_segundo=170)
```
